```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('student_data.csv')
data.info()
```

```python
data['G-avg']=(data.G3+data.G2+data.G1)/3
data['G-avg'].head()
```

```python
plt.hist(data.sex);
plt.title("The number of male and female students")
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=data, x="age", y="G-avg",hue="sex",style="sex")
plt.title("Comparison Between Male and Female Grade");
```

```python
plt.figure(figsize=(10,5))
ax = sns.lineplot(x="absences", y="G-avg", data=data)
```

```python
plt.hist(data.school)
plt.title("Number of students going to diffrent schools");
```

```python
plt.figure(figsize=(10,5))
ax = sns.barplot(x="school", y="G-avg", data=data,hue="sex")
```
